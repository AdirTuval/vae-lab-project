```python
from shapes_dataset_generator import ShapesDatasetGenerator
import matplotlib.pyplot as plt
import numpy as np
```

```python
sdg = ShapesDatasetGenerator(render_config={
    'canvas_size' : 32,
    'background_color' : 'black',
    'sample_color' : 'blue'
    }, sampler_config={
        'type' : 'uniform',
        'params' : [0, 1]
    })
```

```python
samples, sources = sdg.generate(100)
```

```python
for s in np.round(sources[:25] , 2):
    print(s[0])
```

```python
num_images = 25
samples = samples[:num_images]
labels = [f"({round(source[0], 2)}, {round(source[1], 2)})" for source in sources[:num_images]]

print(labels)
# Create a grid layout for the images
num_cols = int(num_images ** 0.5)
num_rows = num_images // num_cols + (1 if num_images % num_cols != 0 else 0)

# Create a figure and axes
fig, axs = plt.subplots(num_rows, num_cols, figsize=(8, 8), gridspec_kw={"wspace": 0.05, "hspace": 0.5})

# Flatten the axes array if it's not already flat
if not isinstance(axs, np.ndarray):
    axs = np.array([axs])

# Plot each image in the grid
for i, ax in enumerate(axs.flat):
    if i < num_images:
        ax.imshow(samples[i])  # Assuming grayscale images
        ax.axis('off')  # Turn off axis
        ax.text(0.5, -0.15, labels[i], fontsize=10, ha='center', transform=ax.transAxes)  # Add text under each image
    else:
        ax.axis('off')  # Turn off empty plots

# Adjust layout
plt.tight_layout()

# Show the grid of images
plt.show()
```
